```python
import numpy as np
import pandas as pd

A = np.array([[2, 4, 6], [1, 3, 5], [7, 9, 11]])
B = np.array([[8, 5, 2], [7, 4, 1], [6, 3, 0]])

#1_sum
sum_AB = A + B

#2_product
product_AB = A * B

#3_dot_product
dot_product = np.dot(A, B.T)

#determinant of A
det_A = np.linalg.det(A)

print("Element-wise Sum:\n", sum_AB)
print("Element-wise Product:\n", product_AB)
print("Dot Product of A and B.^T:\n", dot_product)
print("Determinant of A:", det_A)
```

Problem 4

```python
import numpy as np

# Generate a NumPy array with 10,000 random values
data = np.random.randn(10000)

# Compute mean, median, and standard deviation
mean_val = np.mean(data)
median_val = np.median(data)
std_dev = np.std(data)

# Find min and max values
min_val = np.min(data)
max_val = np.max(data)

# Count values greater than mean
count_greater_than_mean = np.sum(data > mean_val)

# Normalize using Min-Max scaling
normalized_data = (data - min_val) / (max_val - min_val)

print("Mean:", mean_val)
print("Median:", median_val)
print("Standard Deviation:", std_dev)
print("Min:", min_val, "Max:", max_val)
print("Count greater than mean:", count_greater_than_mean)
print("Normalized Data (First 10 values):", normalized_data[:10])
```

Problem 5


```python
temperatures = [28, 32, 35, 31, 29, 30, 33, 36, 27, 25, 34, 30, 29, 37, 38, 26, 31, 35, 33, 32, 36, 34, 29, 28, 27, 35, 32, 30, 31, 33]

# 1. List of temperatures above 32°C
above_32 = [temp for temp in temperatures if temp > 32]

# 2. Classification into "Hot", "Warm", or "Cool"
classified_temps = [(temp, "Hot" if temp > 35 else "Warm" if 30 <= temp <= 35 else "Cool") for temp in temperatures]

# 3. Temperature differences from the monthly average
avg_temp = sum(temperatures) / len(temperatures)
temp_differences = [temp - avg_temp for temp in temperatures]

print("Temperatures above 32°C:", above_32)
print("Classified Temperatures:", classified_temps)
print("Temperature Differences from Average:", temp_differences)
```

```python
df = pd.read_csv('employees_large.csv')
print(df.head())
```

```python
#Average Salary of each Department
avg_salary = df.groupby("Department")["Salary"].mean()
for department, salary in avg_salary.items():
    print(f"Average Salary in {department}: ${salary:.2f}")
```

```python
#Heighest salary
highest_avg_department = avg_salary.idxmax()
print("Department with Highest Average salary: ", highest_avg_department)
```
