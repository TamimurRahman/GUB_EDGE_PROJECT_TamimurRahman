```python
#without comprehension
colors = ['pink','yellow','green','gold','white','gray','green','brown','blue']
newcolor = []
for i in colors:
    if 'e' in i:
      newcolor.append(i)
print(newcolor)
```

```python
#with comprehension
colors = ['pink','yellow','green','gold','white','gray','green','brown','blue','C','C++']
colors.remove(colors[8])
print(colors)
colors.pop(3)
colors.sort()
print(colors)
```
