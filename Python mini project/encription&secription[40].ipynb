```python
#decripted portion main project
import random
import string
def decripted_fun(increpted_sentence):
  word = increpted_sentence.split()
  print(word)
  decriped_list=[]
  new_word = []
  for i in word:
    length = len(i)
    if(length>=3):
      remuve_char = i[:-3]
      temp = remuve_char[len(remuve_char)-1]
      remuve_char=temp+remuve_char
      remuve_char = remuve_char[:-1]
      new_word.append(remuve_char)
    else:
      reverse_words=''.join(reversed(i))
      new_word.append(reverse_words)
  decriped_list=' '.join(new_word)
  print("Decripted sentence is : ",decriped_list)

#encription portion
def encripted_fun(sentence):
  word = sentence.split()
  inceripted_list =[]

  print(word)
  for i in word:
    length=len(i)
    if(length>=3):
     alphabet = list(i)
     temp = i[0]
     alphabet.pop(0)
     alphabet.append(temp)
     letters =  random.choices(string.ascii_lowercase,k= 3)
     alphabet+=letters
     word="".join(alphabet)
    else:
        word = ''.join(reversed(i))

    inceripted_list.append(word)

  print(inceripted_list)
  increpted_sentence = " ".join(inceripted_list)
  print('Increpetaed messeage: ',increpted_sentence)
  decripted_fun(increpted_sentence)


sentence = "Python is a . great programming language"
encripted_fun(sentence)

```
