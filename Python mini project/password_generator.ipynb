```python
import random
import string

def generate_password(length=12):
    if length < 4:  # Ensure password is long enough for security
        return "Password length must be at least 4 characters."

    # Character sets
    letters = string.ascii_letters  # Uppercase and lowercase letters
    digits = string.digits          # Numbers
    symbols = string.punctuation    # Special characters
    print(letters)
    print(digits)
    print(symbols)

    # Ensure password contains at least one of each type
    password = [
        random.choice(letters),
        random.choice(digits),
        random.choice(symbols)
    ]
    print(password)

    # Fill the rest of the password with random characters from all sets
    all_characters = letters + digits + symbols
    print('all carecters: ',all_characters)
    password += random.choices(all_characters, k=length - 3)
    print('password: ',password)
    # Shuffle to ensure randomness
    random.shuffle(password)
    print(password)

    # Convert list to string
    return ''.join(password)

# Example usage
length = int(input("Enter password length: "))
print("Generated password:", generate_password(length))
```
