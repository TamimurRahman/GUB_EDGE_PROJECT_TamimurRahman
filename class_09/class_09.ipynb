add a new item in to  a set

```python
set_a = {10, 2, 3}
num = int(input('Num: '))

# Create set_b correctly
set_b = {num}  # set_b should be a set, not a dictionary
# Add num to set_a using union (or you could use update)
set_a = set_a.union(set_b)  # This returns a new set, so we assign it to set_a

# Alternatively, you could use the update method like this:
# set_a.update(set_b)

print(set_a)
```

```python
set_a.clear()

set_a.add(10)

print(set_a)
```
