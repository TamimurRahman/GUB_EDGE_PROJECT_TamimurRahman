Problem 1) Write a Python Program to Count Total Number of Character with and without Space from a Sentence (Given from User).

```python
value = input('Input string: ')
print('Total length with space: ',len(value))
count = 0
for i in range(0,len(value),1):
  if(value[i]!=' '):
    count +=1
print('Length of the string without space: ',count)
```

Problem 2) Write a Python Program to Display “Even” and “Odd” in a List when it
Finds Even Number and Odd Number, respectively.






```python
li = [1,2,3,10,2,5,6,7,13,20]
newli = []
for i in li:
  if(i%2==0):
    newli.append('True')
  else:
    newli.append('False')
print(newli)
```

Problem 3) Write a Python Program to Create a Dictionary from Given Dictionary,
that Displays Square of Numeric Key.

```python
def square_numeric_keys(original_dict):
  new_dict = {}
  for key,value in original_dict.items():
    if isinstance(key,(int,float)): #check key is numeric
      new_dict[key**2]=value
    else:
      new_dict[key]=value
  return new_dict
original_dict = {1:'one',2:'two','a':'apple',3.5:'three and a half',4:'four'}
print(square_numeric_keys(original_dict))
```

Problem 4) Write a Python Program to Find Min and Max Number from a List.
Then Find Multiplication of Min and Max Number. Then Find Square Root of that
Number to Display to User.

```python
dic={
    'a':'apple',
    'b':'banana',
    'c':'carrot',
    'd':'dragon fruit'

}

while True:
  choice  = input('Input key and press 1 is exit: ')
  print('The fruit name is : ',dic[choice])
  if choice=='1':
    print('Exit')
    break
```
