```python
#list in python:
my_list =[10, 20, 30]
my_list.append(15)  # Inserts 15 at the end of list
print("Unsorted list: ",my_list)
for x in range(0,len(my_list),1):
    min =my_list[x]
    for y in range(x,len(my_list)-1,1):
      if(my_list[x]>my_list[y+1]):
          min=my_list[x]
          my_list[x]=my_list[y+1]
          my_list[y+1]=min
print("sorted list: ",my_list)
```

```python
'''
Generate Odd Numbers From 1 to 20 and Generate Even Numbers from 20 to
1. And then add Both Results to Each Other. Total Result Should Display.
'''
my_list_even = []
my_list_odd = []
my_list_final = []
sum = 0
for x in range(1,21,2):
    my_list_odd.append(x)

for y in range(20,0,-2):
    my_list_even.append(y)
for y in range(0,10,1):
    sum = sum+(my_list_even[y]+my_list_odd[y])
    my_list_final.append((my_list_even[y]+my_list_odd[y]))

print(my_list_final)
print("Sum of all index: ",sum)
```

```python
'''
Get Starting and Ending Numbers from User to Generate a Sequence of
Number. You Should Check at First, Starting Number should be Less Than Ending
Number.
'''
start = int(input("Starting number: "))
end = int(input("Ending number: "))
my_list = []
if(start<end):
    for i in  range(start,end+1,1):
        my_list.append(i)

print(my_list)
```

```python
#using while loop print even or odd sum
even = []
odd =[]
new_even_odd = []
i = 1
j = 20
k=0
sum = 0
while(i<=20):
    odd.append(i)
    even.append(j)
    j = j-2
    i = i+2
    sum = sum+(odd[k]+even[k])
    new_even_odd.append((odd[k]+even[k]))
    k = k+1

print(new_even_odd)
print("Sum: ",sum)

```

```python
# Get two numbers from the user
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))

# Perform logical OR and AND operations
logical_or = num1 | num2  # Bitwise OR
logical_and = num1 & num2  # Bitwise AND

# Display the results
print("Logical OR (|) result:", logical_or)
print("Logical AND (&) result:", logical_and)
```
