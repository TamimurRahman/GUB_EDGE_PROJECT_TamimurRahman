```python
# prompt: need two set union

def find_unique_words(file_path):
  """Finds unique words in a text file.

  Args:
      file_path: The path to the text file.

  Returns:
      A set of unique words.
  """
  unique_words = set()
  try:
    with open(file_path, 'r') as f:
      for line in f:
        words = line.strip().split()
        for word in words:
          unique_words.add(word)
    return unique_words
  except FileNotFoundError:
    print(f"File not found: {file_path}")
    return set()


# Get the file path or content.
file_path = 'your_file.txt'  # Replace with your file name


# Find unique words from the first file.
set1 = find_unique_words(file_path)


# Let's assume another file or string, adjust as needed.
file_path2 = 'another_file.txt'  # Replace with another file or string.
set2 = find_unique_words(file_path2)

# Perform set union
union_of_sets = set1.union(set2)

print("Unique words in the files (Union): ", union_of_sets)
```

```python
# prompt: union between two set

def find_unique_words(text):
  """Finds unique words in a given text."""
  unique_words = set()
  words = text.strip().split()
  for word in words:
    unique_words.add(word)
  return unique_words


# Example usage with strings
text1 = "This is a sample text. This text has some words."
text2 = "Another sample text with some more words."

# Find unique words from the first text.
set1 = find_unique_words(text1)

# Find unique words from the second text.
set2 = find_unique_words(text2)

# Perform set union
union_of_sets = set1.union(set2)

print("Unique words in the texts (Union): ", union_of_sets)
```

```python
# Define two sets
set1 = {1, 2, 3}
set2 = {3, 4, 5}

# Using the union() method
union_set = set1.union(set2)
print("Union using union():", union_set)

# Using the | operator
union_set_operator = set1 | set2
print("Union using | operator:", union_set_operator)
```

```python
# Define complex numbers
a = complex(2, 3)  # 2 + 3i
b = complex(1, 4)  # 1 + 4i

# Basic operations
sum_result = a + b             # Addition
difference_result = a - b      # Subtraction
product_result = a * b         # Multiplication
quotient_result = a / b        # Division

# Magnitude (absolute value) of a complex number
magnitude_a = abs(a)
magnitude_b = abs(b)

# Conjugate of a complex number
conjugate_a = a.conjugate()
conjugate_b = b.conjugate()

# Display results
print("Complex Numbers:")
print("a =", a)
print("b =", b)

print("\nResults:")
print("Sum:", sum_result)
print("Difference:", difference_result)
print("Product:", product_result)
print("Quotient:", quotient_result)
print("Magnitude of a:", magnitude_a)
print("Magnitude of b:", magnitude_b)
print("Conjugate of a:", conjugate_a)
print("Conjugate of b:", conjugate_b)
```

```python
import numpy as np

# Coefficients of the equation: az^2 + bz + c = 0
a = 1  # Coefficient of z^2
b = complex(2, 3)  # Coefficient of z
c = complex(1, -4)  # Constant term

# Using the quadratic formula: z = (-b ± √(b² - 4ac)) / 2a
discriminant = b**2 - 4*a*c
sqrt_discriminant = np.sqrt(discriminant)

# Two possible solutions
z1 = (-b + sqrt_discriminant) / (2*a)
z2 = (-b - sqrt_discriminant) / (2*a)

# Display results
print("Solutions:")
print("z1 =", z1)
print("z2 =", z2)
```
