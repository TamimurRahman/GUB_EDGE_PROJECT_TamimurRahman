```python
# prompt: Suppose  I have an data DNA = [ATCAATCHTTACT] now i count freequency

DNA = "ATCAAGGTCTTACT"
count = 0
frequency = {}
for nucleotide in DNA:
  if nucleotide in frequency:
    frequency[nucleotide] += 1
    count = count+1
  else:
    frequency[nucleotide] = 1
    count = count+1

print(frequency)
print("Count: ",count)
```
