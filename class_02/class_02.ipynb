```python
# prompt: write a pythone code prime number generation

def generate_primes(limit):
  """Generates prime numbers up to a given limit."""
  primes = []
  for num in range(2, limit + 1):
    is_prime = True
    for i in range(2, int(num**0.5) + 1):
      if num % i == 0:
        is_prime = False
        break
    if is_prime:
      primes.append(num)
  return primes

# Example usage:
limit = 50
prime_numbers = generate_primes(limit)
print(f"Prime numbers up to {limit}: {prime_numbers}")
```
