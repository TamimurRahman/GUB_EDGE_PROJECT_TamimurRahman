```python
#Question 02
num1,num2 = map(int,input("Input two numbers: ").split(" "))
print("1.Addition\n2.Difference\n3.product\n4.quotient")
option=int(input("Input your operation: "))
match option:
  case 1:

    print("The sum of two numbers: ",num1+num2)
  case 2:
    print("The difference: ",num1-num2)
  case 3:
    print("The product: ",num1*num2)
  case 4:
    print("The quotient: ",num1//num2)
  case _:
    print("Invalid input!!")
```
